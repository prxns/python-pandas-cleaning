```python
import pandas as pd

# Verify pandas version
print(f"Pandas version: {pd.__version__}")
```

```python
# 1. Load the CSV dataset into a Pandas DataFrame
file_path = 'data/raw/shopping_data.csv'
df = pd.read_csv(file_path)

print("Dataset loaded successfully!")
```

```python
# 2. Explore data
print("--- First 5 Rows ---")
display(df.head())

print("\n--- Last 5 Rows ---")
display(df.tail())

print(f"\nDataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")

print("\n--- Data Types and Non-Null Counts ---")
df.info()
```

```python
# 3. Handle missing values
print("--- Missing Values per Column ---")
print(df.isnull().sum())

df_cleaned = df.dropna(subset=['title', 'final_price']) 

# Fill remaining missing text values (like 'discount', 'seller_name', 'videos') with 'Unknown'
df_cleaned = df_cleaned.fillna('Unknown')

print(f"\nShape after handling missing values: {df_cleaned.shape}")
```

```python
# 4. Perform basic operations
# Filter: Keep only items where the rating is greater than 4.0
df_cleaned = df_cleaned[df_cleaned['rating'] > 4.0]

# Select: View a subset of the actual columns in this dataset
subset_df = df_cleaned[['title', 'final_price', 'rating', 'seller_name']]

print("--- Data Subset (Highly Rated Products) ---")
display(subset_df.head(3))
```

```python
# 5. Remove duplicates 
duplicates_count = df_cleaned.duplicated().sum()
print(f"Found {duplicates_count} duplicate rows.")

df_cleaned = df_cleaned.drop_duplicates()
print(f"Shape after dropping duplicates: {df_cleaned.shape}")
```

```python
# 6. Create a derived column
df_cleaned['product_summary'] = df_cleaned['title'].astype(str) + " (Sold by: " + df_cleaned['seller_name'].astype(str) + ")"

print("Added 'product_summary' column:")
display(df_cleaned[['title', 'seller_name', 'product_summary']].head())
```

```python
# 7. Save the cleaned dataset as a new CSV file
output_path = 'data/processed/cleaned_shopping_data.csv'

# index=False prevents pandas from writing the row numbers into the CSV
df_cleaned.to_csv(output_path, index=False)
print(f"Cleaned dataset saved successfully to {output_path}")
```
